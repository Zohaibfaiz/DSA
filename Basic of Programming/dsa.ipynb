```python
# reverse
# [::-1]
# nums.reversed()
# swap
# left +=1
# right-=1
# nums[l],nums[r]=nums[r],nums[l]

nums=[5,9,8,3,6,7,1,4,2]
l=0
r=(len(nums)-1)
def reve(nums,left,right):
  if left>=right:
    print(nums)
    return
  nums[left],nums[right]=nums[right],nums[left]
  reve(nums,left+1,right-1)
reve(nums,l,r)
```
