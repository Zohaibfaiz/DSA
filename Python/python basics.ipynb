```python
class Mobile:

  def __init__(self,brand,price):
    self.b=brand
    self.p=price
  def discount(self,d):
    self.dis=d
    d=(self.p*self.dis)/100
    self.fp=self.p-d


  def detail(self):
    print("model is:",self.b)
    print("price is:",self.p)
    print("discount is:",self.dis,"%","and discounted price is:",self.fp)

m1=Mobile("Iphone",1250000)
m1.discount(10)
m1.detail()

m2=Mobile("Android",35000)
m2.discount(5)
m2.detail()
```
